```python
import os
import re
import json
import numpy
import networkx as nx
import pandas as pd
```

<h3>Filtering for tweets from 2014 and 2015</h3>

```python
locations = [places for places in os.listdir("profiles")]

print(locations)
tweetsByDate = dict()

for i in locations:
#     print(os.listdir("profiles/"+i))
    
    if(i!="_DS_Store" and i!=".DS_Store"):
        for j in os.listdir("profiles/"+i):
            if j != "_DS_Store" and j!=".DS_Store":
                with open("profiles/"+str(i)+"/"+str(j)) as json_data:
#                     print("profiles/" + i + "/" + j)
                    data = json.load(json_data, strict=False) #list of dictionaries
                    
                    for tweet in data:
                        date = tweet['created_at'].split(" ")[-1]
#                         print(date)
                        if(date == "2014" or date == "2015"):
                            tweetsByDate[tweet['id']] = tweet
                            


```

```python
print(len(tweetsByDate))

print(len(locations))
```

<h3>Filtering for tweets related to the Umbrella Protest</h3>

```python
important_tags = ["occupycentral","unniversalsuffrage","hongkongprotests","pressfreedom","hongkong","hkclassboycott","hkstudentstrike","umbrellarevolution","hongkongautonomy","freehongkong","hkpolicebrutality","standwithhongkong","policestate",]

tweetsByHashtag = dict()

for key in tweetsByDate.keys():
    for i in important_tags:
        for hashtag in tweetsByDate[key]['entities']['hashtags']:
            if hashtag['text'].lower() == i and (key not in tweetsByHashtag.keys()):
                tweetsByHashtag[key] = tweetsByDate[key]    
```

```python
mentions = dict()
quote_dict = dict()
x=set()
for key in tweetsByHashtag.keys():
    #l = []
    tweeter = tweetsByHashtag[key]["user"]["screen_name"]
    x=x.union({tweeter})
    mentioned_users = [i["screen_name"] for i in tweetsByHashtag[key]["entities"]["user_mentions"]]
#     print(tweetsByHashtag[key]['is_quote_status'])
#     print(type(tweetsByHashtag[key]['is_quote_status']))
    if(tweeter not in mentions.keys()):
        mentions[tweeter] = mentioned_users
    else:
        mentions[tweeter].extend(mentioned_users) 
        
    if(tweetsByHashtag[key]['is_quote_status']):
#         print(tweetsByHashtag[key])
        try:
            quoted_user = [tweetsByHashtag[key]["quoted_status"]["user"]["screen_name"]]
            print(quoted_user)
            if tweeter not in quote_dict.keys():
                quote_dict[tweeter] = quoted_user
            else:
                quote_dict[tweeter].extend(quoted_user)
        except:
            continue
```
